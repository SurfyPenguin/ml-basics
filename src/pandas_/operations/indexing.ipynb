# Indexing in pandas

## Index based selection

```python
melbourne_data.iloc[0]
```

Both `iloc` and `loc` are row-first, column-second

```python
melbourne_data.iloc[:, 0]
```

`:` comes from native python, means "everything"

```python
melbourne_data.iloc[:3, 0]
```

```python
melbourne_data.iloc[1:3, 0]
```

Its also possible to pass a list.

```python
melbourne_data.iloc[[0, 1, 3], 0]
```

```python
melbourne_data.iloc[-5, :]
```

### Label-Based Selection

The second paradigm for attribute selection is the one followed by the `loc` operator: __label-based selection__. In this paradigm, it's the data index value, not its position, which matters.

```python
melbourne_data.loc[0, "Address"]
```

```python
melbourne_data.loc[1:10, ["Rooms", "Type", "Price", "Method", "SellerG"]] # row inclusive
```

```python
melbourne_data.loc["Room":"Type"]
```

### Manipulating the Index

Label-based selection derives its power from the labels in the index. Critically, the index we use is not immutable.
The `set_index()` method can be used to set a specific column as index.

```python
melbourne_data.set_index("Address")
```

### Conditional selection

```python
melbourne_data.Type == "h"
```

```python
melbourne_data.loc[melbourne_data.Type == "h"]
```

```python
melbourne_data.loc[(melbourne_data.Method == "S") & (melbourne_data.SellerG == "Biggin")].tail()
```

```python
melbourne_data.loc[melbourne_data.Suburb.isin(["West Footscray", "Prahran"])]
```

```python
melbourne_data.loc[melbourne_data.YearBuilt.notnull()]
```
