## Maps

```python
reviews_points_mean = reviews.points.mean()
reviews.points.map(lambda p: p - reviews_points_mean) #  always returns series
```

`apply()` is the equivalent function method if we want to transform a whole DataFrame by calling a custom method on each row.

```python
def mean_points(row):
    row.points = row.points - reviews_points_mean
    return row

reviews.apply(mean_points, axis='columns')
```

`map()` and `apply()` both return new, Series and DataFrames, respectively.
