# Grouping and Sorting

```python
import pandas as pd

reviews = pd.read_csv("../inputs/winemag-data-130k-v2.csv", index_col=0)
reviews.head()
```

```python
reviews.groupby("points").count()
```

```python
reviews.groupby("points").max()
```

```python
reviews.head(30).groupby("points").sum()
```

## Aggregations

```python
reviews.groupby("points").agg({"points": ["mean", "max", "count", "sum"]})
```

```python
reviews.groupby("points").describe()
```
