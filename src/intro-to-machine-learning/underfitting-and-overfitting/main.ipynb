# Underfitting & Overfitting

* __Underfitting__ is when your model poorly in both training & validation data.
* __Overfitting__ is when your model performs well in training data but poorly in validation data or other new data.

```python
# get the same model lol
import pandas as pd

melbourne_data = pd.read_csv("../inputs/melb_data.csv")

# clean
melbourne_data = melbourne_data.dropna(axis=0)
```

```python
y = melbourne_data.Price

melbourne_features = ["Rooms", "Bathroom", "Landsize", "BuildingArea", "YearBuilt", "Lattitude", "Longtitude"]
X = melbourne_data[melbourne_features]
```

```python
from sklearn.model_selection import train_test_split

# split into training and validation data
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=0)
```

```python
# get minima of absolute error

from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_moa(max_leaf_nodes, train_X, val_X, train_y, val_y):
    melbourne_model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    melbourne_model.fit(train_X, train_y)

    pred_val = melbourne_model.predict(val_X)
    moe = mean_absolute_error(pred_val, val_y)

    return moe
```

```python
for max_leaf_nodes in [5, 50, 500, 1000, 5000]:
    moa = get_moa(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print(f"Max leaf nodes: {max_leaf_nodes} \t\t Mean Absolute Error: {moa}")
```
